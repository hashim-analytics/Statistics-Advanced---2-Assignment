Question 1: What is hypothesis testing in statistics?
Answer:
Hypothesis testing is a statistical method used to make inferences or decisions about a population based on sample data. It involves formulating a hypothesis, collecting data, and then using a statistical test to determine whether the observed data supports or contradicts the hypothesis. Essentially, it helps determine if a claim or assumption about a population parameter is likely to be true or false.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer:

Null Hypothesis (H₀): A statement of no effect or no difference. It represents the default or status quo assumption that we aim to test against.

Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis, representing the effect, difference, or relationship that the researcher expects to find.

Difference:
The null hypothesis assumes no change or effect, while the alternative hypothesis assumes a specific effect or change exists.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
Answer:

Significance Level (α): The probability threshold for rejecting the null hypothesis, commonly set at 0.05 (5%).

Role: It determines how strong the evidence must be to reject the null hypothesis. If the p-value from the test is less than α, we reject the null hypothesis; otherwise, we fail to reject it. A smaller α reduces the risk of a Type I error (false positive).



Question 4: What are Type I and Type II errors? Give examples of each.
Answer:

Type I Error (False Positive): Occurs when we reject the null hypothesis even though it is true.

Example: A new drug is actually ineffective, but a clinical trial concludes that it works.

Type II Error (False Negative): Occurs when we fail to reject the null hypothesis even though it is false.

Example: A new drug is effective, but a clinical trial concludes that it does not work.

Summary: Type I = rejecting true H₀, Type II = failing to reject false H₀.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
Answer:

Z-test: Used when the population variance is known, and the sample size is large (typically n > 30). It uses the standard normal distribution.

Example: Testing the average height of adults in a country when population variance is known.

T-test: Used when the population variance is unknown and/or the sample size is small (n < 30). It uses the Student’s t-distribution.

Example: Testing the mean exam score of a class of 20 students when population variance is unknown.

Key Difference: Z-test assumes known variance and large samples; T-test is for unknown variance and small samples.





```python
# Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
# Answer:


# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Parameters for binomial distribution
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of experiments

# Generate binomial random numbers
binom_data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(binom_data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.
# Answer:

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population mean (H0)
population_mean = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Z-test statistic
z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: Sample mean is not significantly different from population mean.")
```

```python
# Question 8: Simulate data from a normal distribution, calculate the 95% confidence interval, and plot the data.
# Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data
np.random.seed(42)
sample_size = 50
mu = 100        # population mean
sigma = 15      # population standard deviation

data = np.random.normal(mu, sigma, sample_size)

# Sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(sample_size)

# 95% confidence interval
confidence_level = 0.95
z_score = stats.norm.ppf(0.975)  # two-tailed
ci_lower = sample_mean - z_score * se
ci_upper = sample_mean + z_score * se

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot the data
plt.hist(data, bins=10, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()




# Explanation:

# Simulated 50 random values from a normal distribution (mean=100, std=15).

# Calculated the 95% confidence interval for the mean using the Z-score.

# Histogram visualizes data, sample mean, and CI boundaries.
```
